```python
import requests as rq
from bs4 import BeautifulSoup

url = "https://www.dufe.edu.cn/news/news/"
text = rq.get(url).text
sp = BeautifulSoup(text,"html.parser")
find_div = sp.findAll("div",attrs={"class":"col-lg-9 col-md-9 col-sm-9 col-xs-12"})
```

```python
import re
import pandas as pd

news = pd.DataFrame(columns=['新闻标题', '发布时间', '发编单位', '网页链接'])

for div in find_div:
    find_li = div.findAll("li")
    for li in find_li:

        
        find_a = li.findAll("a")
        #print(find_a)

        #网址
        a_tag = find_a[0]  # 获取列表中的第一个 <a> 标签
        href = a_tag.get('href')
        website = "https://www.dufe.edu.cn/"+ href
        print(website)  # 这将输出: content_83811.html
        
        #标题
        for a in find_a:
            title = re.findall(".+",a.string)
            print(title)
           
        #时间
        find_span = li.findAll("span")
        #print(find_span)
        for span in find_span:
            time = re.findall(r"\d{4}\.\d{2}\.\d{2}",span.string)
            print(time)
        
        #单位
        news_link = website
        text1 = rq.get(news_link).text
        sp1 = BeautifulSoup(text1,"html.parser")
        find_p = sp1.findAll("p",attrs={"class":"xw-rel-info"})
        #print(find_p)
        for p in find_p:
            find_span = p.findAll("span")
            #print(find_span)
            publisher = find_span[2].text
            match = re.search(r'单位：(.*?)$', publisher)
            publisher = match.group(1)
            print(publisher)
          

        news.loc[len(news.index)] = [title[0], time[0] , publisher, website]
```

```python
from datetime import datetime
news.sort_values('发布时间', inplace=True)

groups = news.groupby('发布时间')

for date_str, group in groups:
    date_obj = datetime.strptime(date_str, '%Y.%m.%d')
    formatted_date = date_obj.strftime('%Y%m%d')
    filename = formatted_date + '.csv'
    group.to_csv(filename, index=False)
    print(f"文件 '{filename}' 已导出。")
```
